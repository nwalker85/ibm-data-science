```python
canada_map = folium.Map(
    location=[29.891502804973086, -97.96713175002418],
    zoom_start=14,
    tiles='Stamen Terrain'
)
ontario = folium.map.FeatureGroup()

ontario.add_child(
    folium.CircleMarker(
        [29.891502804973086, -97.96713175002418], radius = 5,
        color = "red", fill_color = "Red"
    )
)
canada_map.add_child(ontario)

folium.Marker([29.891502804973086, -97.96713175002418], popup='My House').add_to(canada_map)
canada_map
```
